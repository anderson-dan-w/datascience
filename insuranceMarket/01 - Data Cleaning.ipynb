```python
from __future__ import print_function, division
import pandas as pd
import scipy
from scipy.stats import mode

pd.options.display.max_rows = 11
print("pandas: {}".format(pd.__version__))
print("scipy: {}".format(scipy.__version__))
```

```python
df = pd.read_csv("training.csv")
for col in df.columns:
    uniqcount = df[col].unique().size
    nullcount = df[col].isnull().sum()
    print("{: >12} has {: >3} unique values and {: >3} nulls (type: {})"
          .format(col, uniqcount, nullcount, df[col].dtype))
df.head(5)
```

## Dealing with NaN
---

Sometimes, it's reasonable to just drop any row with `NaN`, but in this case, almost 90% of our rows have `NaN` for `profit`, and almost 10% have `NaN` for `day_of_week`, so it seems like that would miss a lot of important data. Also, we will want to know how to deal with `NaN` rows in our test-prediction set, so we may as well come up with some ideas.

Consider surrounding data and see what makes the most sense.

  - `custAge` - depends how deep we want to get. We can just populate with average `custAge`, or we could do `average-based-on-profession` where we take averages on a per-profession basis, or even just 3 broad categories of `student`, `retired`, and `everything-else`. 
  - `schooling` - "average" doesn't mean anything, but `mode` sounds somewhat reasonable
  - `day_of_week` - same as `schooling`
  - `profit` - Especially for rows where `responded=no`, we can assume the `profit=0`; for rows where `responded=yes`, maybe we could take the average.

```python
averageAge = df.custAge.mean()
df.custAge = df.custAge.fillna(averageAge)

## df[col].mode() seems to include NaN as a possible output,
## whereas scipy.stats.mode doesn't.
## We obviously don't want NaN as the mode because that's what we're replacing
schoolingMode = mode(df.schooling)[0][0]
df.schooling = df.schooling.fillna(schoolingMode)

dayMode = mode(df.day_of_week)[0][0]
df.day_of_week = df.day_of_week.fillna(dayMode)
```

```python
def setProfit(row):
    if pd.isnull(row.profit) and (row.responded == "no"):
        return 0
    return row.profit
df.profit = df.apply(setProfit, axis=1)
```

```python
for col in df.columns:
    uniqcount = df[col].unique().size
    nullcount = df[col].isnull().sum()
    if nullcount > 0:
        print("{: >12} has {: >3} unique values and {: >3} nulls"
              .format(col, uniqcount, nullcount))
```

## Converting to numeric
---

`sklearn` likes numeric data, so it would be helpful to convert columns like `profession` and `schooling` into numeric columns. We *could* map each unique value onto a consecutive number, but then we run the risk of accidentally taking "averages" of something that isn't logically numeric. Instead, we can expand our table (by quite a bit, in some cases) with dummy-columns

```python
nonnumeric = []
for col in df.columns:
    if df[col].dtype == pd.np.object:
        nonnumeric.append(col)
        print(col)
```

```python
for col in nonnumeric:
    dummies = pd.get_dummies(df[col], prefix=col)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop([col], axis=1)
```

```python
print(df.columns)
df.head(5)
```

## Rearrange
---
Just for `sklearn`/reader convenience, let's move our output columns (`responded` and `profit`) to the front.

*minor note* - We `dummy`-fied our `responded` column, but it only had 2 options, `yes` and `no` to begin with; this might've been easier to map, so we'll actually collapse that into a single column again, just keeping `responded_yes`.

```python
columns = df.columns.tolist()
resp, profit = columns.index("responded_yes"), columns.index("profit")
print(resp, profit)
```

```python
reordered = ["responded_yes", "profit"] + columns[:profit] + columns[profit+1 : resp] + columns[resp+1 :]
reordered.remove("responded_no")
df = df[reordered]
df.head(5)
```
