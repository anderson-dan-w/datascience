```python
from __future__ import print_function, division
import pandas as pd
import scipy
import sklearn

from scipy.stats import mode
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor

pd.options.display.max_rows = 11
print("pandas: {}".format(pd.__version__))
print("scipy: {}".format(scipy.__version__))
print("sklearn: {}".format(sklearn.__version__))
```

Need to make sure our testingCandidate table looks the same as what we use for our training - so we will join our tables, and re-do our `NaN` fill-ins.

```python
train = pd.read_csv("training.csv")
## pull out our training dependent variables: respond, profit
trainRespond, trainProfit = train.responded, train.profit
## convert respond from str to boolean; un-NaN profit
trainRespond = trainRespond.map({"yes":1, "no":0})
trainProfit = trainProfit.fillna(0.0)

## remove respond, profit so we match test csv structure
train = train.drop(["responded", "profit"], axis=1)

test = pd.read_csv("testingCandidate.csv", index_col=0)
df = pd.concat([train, test])
df = df.drop(["may_respond", "id"], axis=1)
df
```

```python
averageAge = train.custAge.mean()
df.custAge = df.custAge.fillna(averageAge)

schoolingMode = mode(train.schooling)[0][0]
df.schooling = df.schooling.fillna(schoolingMode)

dayMode = mode(train.day_of_week)[0][0]
df.day_of_week = df.day_of_week.fillna(dayMode)

for col in df.columns:
    if df[col].dtype == pd.np.object:
        dummies = pd.get_dummies(df[col], prefix=col)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop([col], axis=1)

df
```

```python
## overwrite train, test with our new frame
ntrain = len(train)
train = df[:ntrain].values
test = df[ntrain:].values
## now I can create my models on train, knowing test has same shape/ordering for prediction
```

Create and fit the models based on all the training data (and the meta-parameters determined during GridSearch in the Modeling notebook).

```python
## determined from GridSearch in Modeling.ipynb
respondParams = dict(
    max_depth=3,
    max_features=0.25,
    n_estimators=50
)

respondModel = RandomForestClassifier(**respondParams)
respondModel.fit(train, trainRespond)
```

```python
## determined from GridSearch in Modeling.ipynb
profitParams = dict(
    max_depth=5,
    max_features=1.0,
    n_estimators=25
)

profitModel = RandomForestRegressor(**profitParams)
profitModel.fit(train, trainProfit)
```

Now, use the models to predict our test outcomes - for both `may_respond` and `expected_profit`. 

```python
may_respond = respondModel.predict(test)
expected_profit = profitModel.predict(test)

may_respond, expected_profit
```

To fully answer the question, we want to know:

  - if the person is likely to respond; otherwise we probably don't want to market to them
  - if the expected return is more than the `MARKET COST`

That tells us whether or not we want to market; based on those we market to we can predict our expected net profit (taking into account the `MARKET COST`).

```python
MARKET_COST = 30.0

testDf = pd.DataFrame(may_respond)
testDf["expected_profit"] = expected_profit

testDf["should_market"] = testDf.apply(lambda r: (r[0] == 1) & (r[1] > MARKET_COST), axis=1)
testDf["net"] = testDf.apply(lambda r: (0 if r[2] == False else r[1] - MARKET_COST), axis=1)

testDf
```

And then a quick count to get an idea of how many people we should market to, and an expected net profit, just tog et an idea of what this campaign should entail.

```python
print("Should market to {} people, for an expected net profit of ${}"
      .format(testDf.should_market.sum(), int(testDf.net.sum())))
```
