## Data cleaning
---

Helpful to know what hte data looks like, and how we want to deal with nulls. In categorical data, I feel it's less egregious to just treat `NaN` as its own category; that might be a little unfair, but it allows us to keep the data, especially since we only have `< 1000` rows; if we had millions of rows, maybe it's fine to just toss any `NaN`s; YMMV.

So, let's look at our columns, see how many unique values they have (reasonable indicator of whether it's continuous or categorical), and if there are `NaN`s.

```python
for col in df.columns:
    uniqcount = df[col].unique().size
    nullcount = df[col].isnull().sum()
    print("{: >12} has {: >3} unique values and {: >3} nulls"
          .format(col, uniqcount, nullcount))
```

Looking at it, `Name`, `Ticket`, and `Cabin` are either too unique, or too un-populated to really seem helpful, so let's drop them.

```python
df = df.drop(["Name", "Ticket", "Cabin"], axis=1)
```

Comparing some of the above, with actual table data (`df.head()` for example), it seems reasonable that `Sex` is categorical, and `Embarked` is categorical. `SibSp` and `Parch` already seem to be numerical in nature, don't have any nulls, and are either categorical or continuous.

So, let's take our categorical columns and convert them into `sklearn`-acceptable formats, namely, numerical. A category is just a label, so we can arbitrarily assign numbers; we don't want to do anything truly numerical like logistic regression with those columns and prediction, but we can use them in classifiers.

```python
def numerize(df, column):
    uniq = df[column].unique()
    catmap = dict(zip(uniq, range(uniq.size)))
    df[column] = df[column].map(catmap).astype(int)

categorical = ["Sex", "Embarked"]
for col in categorical:
    numerize(df, col)

df[categorical]
```

### Null values

What do we want to do with our remaining nulls? In this case, the only column with `NaN` that still exists (because `numerize` cleaned `NaN` for us), is `Age`. We could drop all those rows, but that's kind of a lot; instead maybe let's convert them to an outlandish number - this will throw off the stats, but hopefully keep it distinct enough, at the very least for viewing purposes. As above, we will need to make sure not to do any `logistic regression` type analyses with `Age` so long as it contains these values.

```python
df.Age = df.Age.fillna(999)
```

Now we can check if we have a perfectly rectangular, all-numeric, non-null array:

```python
print(df.info())
df.describe()
```

## Fill NaN with Surrounding Data

Sometimes, it makes more sense to just fill in missing data with some appropriate value based on the nearby data.

 - To fill in `Age`, let's replace the `NaN`s with the `mean`, though we could also argue whether the `median` might be better.
 - To fill in `Embarked`, where the `mean`/`median` doesn't really make sense, we can fill it in with the `mode`, since `Embarked` is categorical

```python
## read in again to start fresh
df = pd.read_csv("data/train.csv")
```

```python
from scipy.stats import mode
## fill age with mean
age_mean = df.Age.mean()
df.Age = df.Age.fillna(age_mean)

## fill embarked with mode
embarked_mode = mode(df.Embarked)[0][0]
df.Embarked = df.Embarked.fillna(embarked_mode)

## drop don't-cares (as before)
df = df.drop(["Name", "Ticket", "Cabin"], axis=1)
## re-numerize categorical columns (as before)
for column in categorical:
    numerize(df, column)
```

## Dummy variables

For categorical variables that don't really have an ordering, it would be nice to not submit numerical-seeming data to these algorithms. Specifically, `Embarked` had values `{C, Q, S}` that are now `{1, 2, 3}`, but that implies an inherent ordering when there really isn't any. Let's convert those into dummies

```python
## read in again to start fresh
df = pd.read_csv("data/train.csv")

## fill age with mean
age_mean = df.Age.mean()
df.Age = df.Age.fillna(age_mean)

## set fare based on Pclass
fare_by_pclass = df.pivot_table(
    values="Fare", index="Pclass", aggfunc=np.mean)
def fareFunc(row):
    if pd.isnull(row["Fare"]) or 0.0 == row["Fare"]:
        return fare_by_pclass[row["Pclass"]]
    return row["Fare"]
df.Fare = df[["Fare", "Pclass"]].apply(fareFunc, axis=1)

## numerize Sex
numerize(df, "Sex")

## drop don't-cares (as before)
df = df.drop(["Name", "Ticket", "Cabin"], axis=1)

## fill embarked with mode
embarked_mode = mode(df.Embarked)[0][0]
df.Embarked = df.Embarked.fillna(embarked_mode)
```
